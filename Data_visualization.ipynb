# Titanic Data Visualization



## 1. Passenger Class Distribution
The bar chart below shows the number of passengers in each class. Most passengers were in 3rd class.



```python
ship.survived.value_counts().plot(kind="bar")
```

## 2. Survival Rate by Gender
The chart illustrates that females had a higher survival rate than males.



```python
sns.barplot(x="sex",y="survived",hue="who",data=ship)
```

```python
ship.columns
```

```python
# scattered plot

plt.scatter(x="fare", y="who", c='age',data=ship, alpha=1.0)
```

```python
sns.barplot(x="who", y="fare", hue='pclass',data=ship )
```

```python

sns.scatterplot(x="age", y="fare", hue='pclass',data=ship, )
```

```python
ship.columns
```

```python
sns.pairplot(ship)
```

## 3. Age Distribution
Histogram shows most passengers were aged between 20-40 years.


```python
ship.hist(bins=10,figsize=(10,10),grid=False)
```

```python
ship.columns
```

```python
sns.boxplot(x="sex",y="fare",data=ship,hue="who" )
```

```python
sns.boxenplot(x="sex",y="fare",data=ship,hue="who" )
```

```python
g=sns.FacetGrid(ship,hue="survived",col="pclass",margin_titles=True,
              palette={1:"seagreen",0:"gray"})
g=g.map(plt.scatter,"fare","age",edgecolor="b").add_legend();
```

```python
sns.stripplot(x="sex",y="fare",hue="class",data=ship)
```

> correlation is from -1 to 1.

```python
sns.violinplot(x="sex",y="fare",hue="class",data=ship)
```

# Summary Insights
- Females had higher survival rates than males.
- Passengers in 1st class had higher survival rates compared to 3rd class.
- Most passengers were aged 20-40 years.
- Fare and class are slightly correlated, indicating higher fares in 1st class.
These insights could be used for understanding historical travel patterns or planning safety strategies.
